# DATA

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_names = ["ctr_15.csv", "ctr_16.csv", "ctr_17.csv", "ctr_18.csv", "ctr_19.csv", "ctr_20.csv", "ctr_21.csv"]
# Read and concatenate all the CSV files for the training data
train_data = pd.concat([pd.read_csv(file) for file in file_names])
```

### Handle Missing Data

```python
# Check for missing data
missing_data = train_data.isnull().sum()
print("Missing data in training set:")
print(missing_data[missing_data > 0])
```

### Exploring Unique Values in Categorical Variables

```python
# List of categorical columns (based on the provided data fields)
categorical_columns = [
    'action_categorical_0', 'action_categorical_1', 'action_categorical_2', 
    'action_categorical_3', 'action_categorical_4', 'action_categorical_5', 
    'action_categorical_6', 'action_categorical_7', 'auction_categorical_0', 
    'auction_categorical_1', 'auction_categorical_2', 'auction_categorical_3', 
    'auction_categorical_4', 'auction_categorical_5', 'auction_categorical_6', 
    'auction_categorical_7', 'auction_categorical_8', 'auction_categorical_9', 
    'auction_categorical_10', 'auction_categorical_11', 'auction_categorical_12', 
    'creative_categorical_0', 'creative_categorical_1', 'creative_categorical_2', 
    'creative_categorical_3', 'creative_categorical_4', 'creative_categorical_5', 
    'creative_categorical_6', 'creative_categorical_7', 'creative_categorical_8', 
    'creative_categorical_9', 'creative_categorical_10', 'creative_categorical_11', 
    'creative_categorical_12', 'device_id_type', 'gender'
]

# Display unique values for each categorical column in the training data
for col in categorical_columns:
    print(f"Unique values in {col}: {train_data[col].nunique()}")
    print(train_data[col].unique())  #[:10])  # Show first 10 unique values for brevity
    print()
```

### Correlation Between Numerical Variables

```python
# List of numerical columns
numerical_columns = [
    'auction_time', 'auction_age', 'auction_bidfloor', 'creative_height', 
    'creative_width', 'timezone_offset'
]

# Calculate correlation matrix
correlation_matrix = train_data[numerical_columns].corr()

# Plot the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features in Training Data')
plt.show()
```

### Analyze Boolean Variables

```python
# List of boolean columns
boolean_columns = ['auction_boolean_0', 'auction_boolean_1', 'auction_boolean_2', 'has_video']

# Calculate mean of the target variable (Label) for each boolean feature
for col in boolean_columns:
    mean_label_by_boolean = train_data.groupby(col)['Label'].mean()
    print(f"Mean 'Label' for {col}:")
    print(mean_label_by_boolean)
    print()
```

### Analyze the Distribution of List Columns

```python
# Example analysis for list columns
list_columns = ['action_list_0', 'action_list_1', 'action_list_2', 'auction_list_0']

# Calculate the number of unique values in the lists and their distribution
for col in list_columns:
    train_data[col] = train_data[col].apply(lambda x: len(str(x).split(',')))
    print(f"Distribution of list lengths in {col}:")
    print(train_data[col].value_counts())
    print()
```

### Visualize the Target Distribution

```python
# Plot the distribution of the target variable
sns.countplot(data=train_data, x='Label')
plt.title('Distribution of Target Variable in Training Data')
plt.show()
```
